## Basic Setup & First Visuals (step-by-step)

```python
# Environment & imports ‚Äî run this first


# Run these in a fresh Jupyter cell (or at start of script)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Optional: prettier inline plots in Jupyter
%matplotlib inline

# Load a sample dataset for examples
tips = sns.load_dataset("tips")

#Use sns.set_theme() so every plot looks professional without extra effort

# Recommended theme & default rc settings
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams['figure.figsize'] = (8, 4)   # default figure size (width, height)

"""
style="whitegrid" ‚Üí subtle grid and clean background (great for teaching)

context="notebook" ‚Üí font sizes appropriate for notebooks; use talk for presentations

rcParams['figure.figsize'] ‚Üí sets a comfortable default chart size
"""


#Shows distribution of a numeric column.

plt.figure()  # clear a new figure
sns.histplot(data=tips, x="total_bill", bins=20, kde=False)
plt.title("Histogram: Total Bill")
plt.xlabel("Total bill ($)")
plt.ylabel("Count")
plt.show()
```

```python
# KDE smooths the histogram ‚Äî useful to see the shape without the bin noise

plt.figure()
sns.histplot(data=tips, x="total_bill", bins=30, kde=True)
plt.title("Total Bill with KDE overlay")
plt.show()
```

```python
# Density plot only (kdeplot)
#For few rows KDE can be misleading ‚Äî good to explain sample size effects.

plt.figure()
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.title("KDE: Total Bill")
plt.show()
```

```python
# Scatter plot (scatterplot) ‚Äî relationship between two numerics

plt.figure()
"""
hue colors points by category (adds legend)

style changes marker shape (good when printing B/W)

s controls marker size
"""
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="sex", s=60)
plt.title("Tip vs Total Bill (hue=time, style=sex)")
plt.xlabel("Total bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Time / Sex", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.show()
```

```python
# Box plot (boxplot) ‚Äî compare distribution by category
plt.figure()
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Boxplot: Total Bill by Day and Sex")
plt.xlabel("Day of week")
plt.ylabel("Total bill ($)")
plt.show()
```

```python
# Bar plot (barplot) ‚Äî aggregate statistic per category

plt.figure()
sns.barplot(data=tips, x="day", y="total_bill", estimator='mean', ci=95)
plt.title("Average Total Bill by Day (95% CI)")
plt.ylabel("Average total bill ($)")
plt.show()

"""
estimator='mean' is default; you can use sum, np.median, etc.

ci=95 shows 95% confidence interval (set ci=None to hide)
"""
```

```python
# Count plot (countplot) ‚Äî frequency of categories

plt.figure()
sns.countplot(data=tips, x="day", hue="time")
plt.title("Count of Bills by Day and Time")
plt.show()

#Use case: quick frequency checks (good before modeling).
```

```python
# Quick layout & saving figures
# adjust layout so legend/title don't overlap
plt.tight_layout()

# save last figure
plt.savefig("myplot.png", dpi=300, bbox_inches="tight")
```

```python
# Box Plot ‚Äî "The OG of distribution comparison"

import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 4))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="Set2")
plt.title("Box Plot: Total Bill by Day & Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.legend(title="Sex")
plt.show()
print(
    """
    Box center line ‚Üí median

Box height ‚Üí IQR (25th to 75th percentile)

Whiskers ‚Üí range of ‚Äúnormal‚Äù data

Dots outside ‚Üí outliers

Common Q: ‚ÄúWhy are some dots outside?‚Äù
üëâ Because those are bills that are way higher or lower than most others.
"""
)
```

## üéª 2Ô∏è‚É£ Violin Plot ‚Äî ‚ÄúBox plot meets distribution curve‚Äù

    A violin plot gives more detail than a boxplot ‚Äî it shows the data density as a smooth curve.

```python
plt.figure(figsize=(8, 4))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True, inner="quartile")
plt.title("Violin Plot: Total Bill by Day & Gender")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()

print(

    """
    split=True ‚Üí draws male & female violins side by side (within one shape)

inner="quartile" ‚Üí draws lines for 25th, 50th, and 75th percentiles

Violin width = data density (fatter = more records in that range)
"""
)
```

#  üåà 3Ô∏è‚É£ Strip Plot ‚Äî "Each dot is a real data point"
## ‚öôÔ∏è Key points:

Each dot = one observation (raw data visibility)

jitter=True ‚Üí spreads points to avoid overlap

dodge=True ‚Üí separates hues side-by-side for clarity

Use this when students should see real individual data ‚Äî not just summaries.

```python
plt.figure(figsize=(8, 4))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", jitter=True, dodge=True)
plt.title("Strip Plot: Every Bill by Day & Gender")
plt.show()
```

# 4Ô∏è‚É£ Swarm Plot ‚Äî ‚ÄúOrganized strip plot (no overlapping!)‚Äù


## ‚ú® Explain:

Similar to stripplot, but smarter: points don‚Äôt overlap.

Visually stunning + honest representation.

Best used for small to medium datasets (can get slow on large ones).



```python
plt.figure(figsize=(8, 4))
sns.swarmplot(data=tips, x="day", y="tip", hue="sex")
plt.title("Swarm Plot: Tips by Day & Gender")
plt.show()
```

# üßÉ 5Ô∏è‚É£ Combining Categorical Plots (Pro Tip Combo)

    Combine violinplot + stripplot (or swarmplot) to show distribution and actual data points.

```python
plt.figure(figsize=(8, 4))
sns.violinplot(data=tips, x="day", y="total_bill", inner=None, color="lightgray")
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, jitter=True, palette="Set2")
plt.title("Combined Plot: Violin + Strip (Total Bill by Day & Gender)")
plt.show()
```

# Seaborn Relational Plots (Scatter, Line, & relplot Mastery)
## 1Ô∏è‚É£ Scatter Plot ‚Äî The ‚Äúclassic‚Äù relationship plot

```python
print(
    """
    Each point = one record (a meal)

x = total bill, y = tip

hue = color by category (male/female)

style = different markers (lunch/dinner)

size = bubble size based on party size
"""
)


import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", size="size", palette="coolwarm")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

# üåà 2Ô∏è‚É£ Line Plot ‚Äî Showing trends over time or sequence
    Breakdown:

    Great for time series or continuous data.
    
    Automatically adds confidence intervals (the shaded area).
    
    hue and style let you compare multiple lines in one chart.
    ‚ÄúThe shaded region represents uncertainty ‚Äî think of it like the ‚Äòmargin of error‚Äô in surveys.‚Äù

```python
# Load seaborn dataset
fmri = sns.load_dataset("fmri")

plt.figure(figsize=(8, 5))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot: fMRI Signal Over Time")
plt.xlabel("Time Point")
plt.ylabel("Signal Intensity")
plt.show()
```

# relplot() ‚Äî The ‚Äúcombo" of relational plots

    relplot() = a figure-level function, unlike scatterplot and lineplot (which are axes-level).
    
    That means you can:
    
    Add subplots easily.
    
    Control overall layout.
    
    Plot relationships across multiple groups

```python
print(
    """
    Explanation:

col="time" ‚Üí creates one subplot for Lunch, another for Dinner.

Each plot shows the relationship between bill & tip.

hue="day" adds color to distinguish weekdays.
"""
)


sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    col="time",
    kind="scatter",
    height=4,
    aspect=1
)
plt.show()
```

# üí• Example 2: Line relplot
# ‚ú® Output: Multiple line charts for each brain region, automatically arranged side by side.

```python
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    kind="line",
    col="region",
    height=4,
    aspect=1.2
)
plt.show()
```

## üé® 4Ô∏è‚É£ Styling & Customizing Your Plots
    You can control the overall look of all seaborn plots using themes:

```python
sns.set_style("whitegrid")  # options: white, dark, ticks, whitegrid, darkgrid
sns.set_context("talk")     # adjusts font size: paper, notebook, talk, poster
sns.set_palette("coolwarm") # sets color theme globally

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", s=100)
plt.title("Stylish Scatter Plot with Custom Theme")
plt.show()
```

# Seaborn Distribution Plots (Understanding Data Shapes)
    to distributions ‚Äî
    ‚ÄúHow are values spread out?‚Äù
    ‚ÄúWhat‚Äôs common, rare, or extreme?‚Äù
    
    Seaborn makes it super easy to visualize that.

# üéØ 1Ô∏è‚É£ Histogram ‚Äî sns.histplot()

    This shows how data values fall into different bins.

```python
print(
    """Key Points:

bins controls how many bars (intervals).

Each bar = number of observations in that range.

Great for seeing skewness or outliers."""
)

import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

plt.figure(figsize=(8,5))
sns.histplot(data=tips, x="total_bill", bins=20, kde=False, color="skyblue")
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Add KDE (Smooth Curve) ‚Äî kde=True
"""
KDE = Kernel Density Estimate

It‚Äôs a smooth version of the histogram ‚Äî showing the probability density.
(Think of it like: how ‚Äúdense‚Äù data points are around certain values.)
"""
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="orchid")
plt.title("Histogram + KDE")
plt.show()
```

# Standalone KDE Plot ‚Äî sns.kdeplot()

    When to use:

    When you want to focus on shape rather than count.
    
    Especially useful for comparing multiple distributions.

```python
sns.kdeplot(data=tips, x="total_bill", fill=True, color="teal")
plt.title("KDE Plot of Total Bill")
plt.show()
```

# ü•á 4Ô∏è‚É£ Multiple Distributions ‚Äî hue & multiple
# üîç What happens:
    
    multiple="stack" ‚Üí stacks bars
    
    multiple="dodge" ‚Üí separates them side by side
    
    multiple="fill" ‚Üí normalizes (percentage form)

```python
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    multiple="stack",   # or "dodge", "fill"
    bins=20,
    kde=True
)
plt.title("Distribution by Gender")
plt.show()
```

# 5Ô∏è‚É£ Pairplot ‚Äî Relationship + Distribution Hybrid

    This is one of Seaborn‚Äôs signature features 

```python


print(
    """
    üí° What‚Äôs happening:

It automatically creates pairwise scatter plots for all numeric columns.

On the diagonal: histograms or KDE plots.

Off-diagonal: scatter plots showing relationships.

‚ö° Best use: quick data overview before modeling."""
)

sns.pairplot(tips, hue="sex", diag_kind="kde")
plt.show()
```

```python
# Rug Plot ‚Äî Tiny but Insightful --Each little red tick (rug) = one data point ‚Äî helps visualize data density.

sns.kdeplot(data=tips, x="total_bill", fill=True)
sns.rugplot(data=tips, x="total_bill", color="red")
plt.show()
```

# Regression & Statistical Plots in Seaborn
# üéØ 1Ô∏è‚É£ sns.regplot() ‚Äî The Regression Line Plot

    This is Seaborn‚Äôs simplest ‚Äúrelationship + trend‚Äù chart.
    It automatically fits a linear regression line.

```python

"""
Concept:

Scatter points = actual data.

Line = best fit line (linear regression).

It‚Äôs the visual ‚Äúpredictive relationship.‚Äù
"""

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.regplot(x="total_bill", y="tip", data=tips, color="mediumseagreen")
plt.title("Regression between Total Bill and Tip")
plt.show()
```

# ‚öôÔ∏è 2Ô∏è‚É£ Add Confidence Interval ‚Äî ci parameter
# Meaning:

The shaded area shows the 95% confidence interval of the regression ‚Äî
It means ‚Äúwe‚Äôre 95% sure the true line lies here.‚Äù

You can set ci=None to remove it for a cleaner look.

```python
sns.regplot(x="total_bill", y="tip", data=tips, ci=95, color="coral")
plt.title("95% Confidence Interval for Regression Line")
plt.show()
```

# Turn Off the Line ‚Äî fit_reg=False

    Just want the scatter points? Easy.
    üí° Great for comparing how the trend looks with vs without the regression

```python
sns.regplot(x="total_bill", y="tip", data=tips, fit_reg=False)
plt.title("Only Scatter Points, No Regression Line")
plt.show()
```

# 4Ô∏è‚É£ Adding Categories with hue using sns.lmplot()

lmplot() is like regplot() but with more flexibility ‚Äî
you can add categories, split plots, or map multiple regressions easily.

üîç What‚Äôs happening:

Each hue category gets its own regression line.

Helps visually compare behavior across groups.

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Regression by Gender")
plt.show()
```

# 5Ô∏è‚É£ Multiple Facets ‚Äî col and row

Use lmplot() to split your dataset into mini-panels (grids).

üí• This automatically creates multiple plots to show differences across subgroups.
It‚Äôs amazing for visual storytelling ‚Äî very GenAI-demo worthy.

```python
sns.lmplot(x="total_bill", y="tip", col="sex", row="smoker", data=tips)
plt.show()
```

#  6Ô∏è‚É£ Non-linear Relationships ‚Äî order parameter
    üé¢ Adds a curved line instead of a straight one (good for complex trends).
    order=2 = quadratic, order=3 = cubic, etc.

```python
sns.regplot(x="total_bill", y="tip", data=tips, order=2, color="tomato")
plt.title("Polynomial Regression (Order = 2)")
plt.show()
```

# üìä 7Ô∏è‚É£ Logistic Regression ‚Äî For Classification-like Data
    
    If your Y variable is binary (like 0 or 1, True/False), Seaborn can fit a logistic regression curve.

# 2Ô∏è‚É£ Bar Plot ‚Äî Mean Value of Each Category

Shows average (mean) value per category.

```python
print(
    """
    Tip:
Bar plots automatically calculate the mean and 95% confidence interval.
"""
)

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill per Day")
plt.show()
```
